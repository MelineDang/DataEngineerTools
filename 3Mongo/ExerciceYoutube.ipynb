```python
import pandas as pd
import pymongo
import json
```

```python
client = pymongo.MongoClient()
database = client['exercices']
collection = database['youtube']
```

```python
df_youtube = pd.read_csv("./data/USvideos.csv")
df_youtube.head()
```

```python
df_category = pd.read_json("./data/US_category_id.json")
df_category['ID'] = df_category.index
df_category.head()
```

```python
df_youtube = pd.merge(df_youtube, df_category, left_on='category_id', right_on='ID')
df_youtube.head()
```

## Question 0

### Netoyer les données

```python
#Type de nos variables
print(df_youtube.dtypes)
```

```python
# Taille de notre jeu de données
print(df_youtube.shape)
# Nous regardons combien nous avons de valeur NaN par colonnes
print(df_youtube.isna().sum())
```

```python
# Nous supprimons les colonnes qui ne sont pas utiles pour répondre aux questions.
df_youtube = df_youtube.drop(columns=["dislikes","comment_total","thumbnail_link","date","ID"])
df_youtube.head()
```

```python
# Nous découpons les tags en listes
df_youtube['tags'] = df_youtube['tags'].str.split("|")
df_youtube.head()
```

### Importer les données

```python
df = df_youtube.to_json(orient="records")
```

```python
df = json.loads(df)
```

```python
collection.insert_many(df)
```

## Question 4
Récupérer les vidéos les plus vues.

```python
# Récupération des vidéos les plus vues.
# Nous trions par views, et prenons les 15 premières.
cur = collection.find().sort("views",-1).limit(15)
list(cur) 
```

## Question 5
Compter le nombre moyen de vues en fonction de la catégorie.

```python
# Comptage du nombre moyen de vuues en fonction de la catégorie.
# Nous regroupons par catégorie, et nous faisons ensuite la moyenne des vues.
cur = collection.aggregate([{"$group" : {"_id" : "$category_id", "moyenne nombre vues" : {"$avg" : "$views"}}}])
list(cur)
```

## Question 6 
Récupérer les chaines Youtube avec la plus grande moyenne de likes.

```python
# Récupération des chaines YouTube avant les plus grandes moyennes de likes.
cur = collection.aggregate([{"$group" : {"_id" : "$channel_title", "chaines Youtube avec la plus grande moyenne de likes" : {"$avg" : "$likes"}}},  { "$sort" : { "chaines Youtube avec la plus grande moyenne de likes" : -1 } }])
list(cur)
```
